```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
train = pd.read_csv('training_set_features.csv')
train_out = pd.read_csv('training_set_labels.csv')
test = pd.read_csv('test_set_features.csv')
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
train = pd.read_csv('training_set_features.csv')
train = train.merge(train_out[['respondent_id', 'xyz_vaccine']], on='respondent_id', how='left')
train = train.merge(train_out[['respondent_id', 'seasonal_vaccine']], on='respondent_id', how='left')
train
```

```python
fig = px.histogram(train, 
                   x='xyz_concern', 
                   color='xyz_vaccine', 
                   color_discrete_sequence=['green', 'grey'])
fig.update_layout(bargap=0.1)
fig.show()
```

```python
percentage_df = train.groupby('xyz_concern')['xyz_vaccine'].value_counts(normalize = True).mul(100).rename('Percentage').reset_index()
percentage_df
```

```python
MSA_code = {'MSA, Not Principle City': 2, 'MSA, Principle City': 1, 'Non-MSA': 0}
train['MSA_code'] = train.census_msa.map(MSA_code)
train.xyz_vaccine.corr(train.MSA_code)
```

```python
sex_code = {'Female': 1, 'Male': 0}
train['sex_code'] = train.sex.map(sex_code)
train.xyz_vaccine.corr(train.sex_code)
```

```python
corr_df=(train.select_dtypes(include=[np.number])).corr()
new_df = corr_df[corr_df['xyz_vaccine'] >= 0.1]
new_df
```

```python
corr_df=(train.select_dtypes(include=[np.number])).corr()
new_df1 = corr_df[corr_df['seasonal_vaccine'] >= 0.1]
new_df1
```

```python
relev_cols = ['xyz_concern', 'xyz_knowledge', 'doctor_recc_xyz',
       'doctor_recc_seasonal', 'health_worker', 'health_insurance',
       'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_seas_vacc_effective', 'opinion_seas_risk']
```

```python
relev_cols1 = ['xyz_concern', 'xyz_knowledge', 'behavioral_wash_hands','behavioral_touch_face','chronic_med_condition', 'doctor_recc_xyz',
       'doctor_recc_seasonal', 'health_worker', 'health_insurance',
       'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_seas_vacc_effective', 'opinion_seas_risk']
```

```python
train['xyz_concern'] = train['xyz_concern'].fillna(train['xyz_concern'].mean())
train['xyz_knowledge'] = train['xyz_knowledge'].fillna(train['xyz_knowledge'].mean())
train['doctor_recc_xyz'] = train['doctor_recc_xyz'].fillna(train['doctor_recc_xyz'].mean())
train['doctor_recc_seasonal'] = train['doctor_recc_seasonal'].fillna(train['doctor_recc_seasonal'].mean())
train['health_worker'] = train['health_worker'].fillna(train['health_worker'].mean())
train['health_insurance'] = train['health_insurance'].fillna(train['health_insurance'].mean())
train['opinion_xyz_vacc_effective'] = train['opinion_xyz_vacc_effective'].fillna(train['opinion_xyz_vacc_effective'].mean())
train['opinion_xyz_risk'] = train['opinion_xyz_risk'].fillna(train['opinion_xyz_risk'].mean())
train['opinion_seas_vacc_effective'] = train['opinion_seas_vacc_effective'].fillna(train['opinion_seas_vacc_effective'].mean())
train['opinion_seas_risk'] = train['opinion_seas_risk'].fillna(train['opinion_seas_risk'].mean())
train['behavioral_wash_hands'] = train['behavioral_wash_hands'].fillna(train['behavioral_wash_hands'].mean())
train['behavioral_touch_face'] = train['behavioral_touch_face'].fillna(train['behavioral_touch_face'].mean())
train['chronic_med_condition'] = train['chronic_med_condition'].fillna(train['chronic_med_condition'].mean())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train[relev_cols])
scaled_inputs = scaler.transform(train[relev_cols])
scaled_inputs
```

```python
scaler.fit(train[relev_cols1])
scaled_inputs1 = scaler.transform(train[relev_cols1])
scaled_inputs1
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(scaled_inputs, train['xyz_vaccine'])
predcitions = model.predict_proba(scaled_inputs)[:,-1]
x = predcitions.astype(float)
x
```

```python
scaler.fit(test[relev_cols])
scaled_test_inputs = scaler.transform(test[relev_cols])
scaled_test_inputs
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
scaled_test_inputs = imputer.fit_transform(scaled_test_inputs)
```

```python
predcitions2 = model.predict_proba(scaled_test_inputs)[:,-1]
x1 = predcitions2.astype(float)
x1
```

```python
model1 = LogisticRegression().fit(scaled_inputs1, train.seasonal_vaccine)
predcitions1 = model1.predict_proba(scaled_inputs1)[:,-1]
y = predcitions1.astype(float)
y
```

```python
scaler.fit(test[relev_cols1])
scaled_test_inputs1 = scaler.transform(test[relev_cols1])
scaled_test_inputs1 = imputer.fit_transform(scaled_test_inputs1)
scaled_test_inputs1
```

```python
predcitions3 = model1.predict_proba(scaled_test_inputs1)[:,-1]
y1 = predcitions3.astype(float)
y1
```

```python
import pandas as pd

# Create a dictionary with the data
data = {
   'respondent_id': train['respondent_id'],
   'x': x,
   'y': y
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('data.csv', index=False)
```
